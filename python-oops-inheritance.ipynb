```python
# Multilevel inheritance

class GrandFather:
    grandfather_prop = "Grandfather's Property!"

class Father(GrandFather):
    father_prop = "Father's Property!"

class Child(Father):
    child_prop = "Child's Property!"


# Create object outside the class
obj = Child()

# Access properties
print(obj.grandfather_prop)
print(obj.father_prop)
print(obj.child_prop)
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age


class Employee(Person):
    def __init__(self, name, age, dep):
        super().__init__(name, age)
        self.dep = dep   # fixed variable name


class Manager(Employee):
    def __init__(self, name, age, dep, team_size):
        super().__init__(name, age, dep)
        self.team_size = team_size

    def details(self):
        print(f'Name: {self.name}')
        print(f'Age: {self.age}')
        print(f'Department: {self.dep}')
        print(f'Team Size: {self.team_size}')


# Create object outside the class
manager = Manager('Himanshu', 24, 'Machine Learning', 50)
manager.details()
```

```python
#Multiple Inhertance -> In this inhertance basically there more than one parent
# of the derived class or the chil class class A, Class B, Class(A,B)
class CngCar:
    prop = 'This car will run with CNG'

    def display(self):
        print('This is the CNG car property!')


class PetrolCar:
    petrol_prop = 'This car will run with petrol'

    def display(self):
        print('This is the Petrol car property!')


class HybridCar(CngCar, PetrolCar):
    pass


car = HybridCar()

car.display()   # Calls method based on MRO
print(HybridCar.__mro__)

# Method resolution order (MRO)
# Method resolution order defines the or decides how the method of the differnt
# class would run..
# A - Display
# B - Dispaly
# c(A, B)
# c - obj -> obj.display -> here class A display method is called
# class - C, Class A, class B
```

```python
class Students:
    college = 'IIT'

    def __init__(self, name):
        self.name = name

    @classmethod
    def change_college(cls, new_college):
        cls.college = new_college


# Creating object
obj = Students('Naman')

# Changing college using class method
obj.change_college('IIT Delhi')

# Printing result
print(obj.college)
```
